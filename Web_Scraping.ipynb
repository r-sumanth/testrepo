# scraping websites with multiple pages

discudemy.com is a website which provides links for courses on udemy that are free of cost after applying coupon

lets scape the site and get the course related to Marketing

```python
import requests
import bs4
```

```python
soup=bs4.BeautifulSoup(requests.get('https://www.discudemy.com/language/English/1').text,'lxml')
```

```python
soup
```

```python
soup.select('.header')
```

```python
len(soup.select('.header'))
```

```python
soup.select('.header')[3].text.replace('\n','')
```

```python
soup.select('.header')[3]('a')[0]['href']
```

```python
import pdb
```

```python
required_keywords=['Marketing','Ads','Advertising']
free_courses={}

base_url="https://www.discudemy.com/language/English/{}"

for n in range(1,11):
    scrape_url=base_url.format(n)
    soup=bs4.BeautifulSoup(requests.get(scrape_url).text,'lxml')
    courses=soup.select('.header')

    for course in courses:
        for keyword in required_keywords:
            if keyword in str(course):
                free_courses.update({course.text.replace('\n',''):course('a')[0]['href']})

print('Free courses avalibale are: ')
for course,link in free_courses.items():
    print(f"\n{course} : {link}")
```
